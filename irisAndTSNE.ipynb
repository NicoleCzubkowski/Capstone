### normalize (standardize) the data

```r
irisN = data.frame(iris)
```

```r
normalize <- function(var){
  return ((var-min(var))/(max(var)-min(var)))
}
```

```r
irisN$Sepal.Length = normalize(iris$Sepal.Length)
irisN$Sepal.Width = normalize(iris$Sepal.Width)
irisN$Petal.Length = normalize(iris$Petal.Length)
irisN$Petal.Width = normalize(iris$Petal.Width)
```

### hierarchical clustering

Distance metric:

$ |a-b|_2 = \sqrt{\sum_i (a_i-b_i)^2} $ 

linkage criteria:

$ \frac{1}{|A|\cdot|B|} \sum_{a \in A }\sum_{ b \in B} d(a,b) $

```r
iris.dist = dist(iris[,c(-5)], method = "euclidean")
```

```r
iris.hclust.fit = hclust(iris.dist, method="ward")
```

```r
plot(iris.hclust.fit)
```

```r
plot(iris.hclust.fit)
groups = cutree(iris.hclust.fit, k=3)
rect.hclust(iris.hclust.fit, k=3, border="red")
```

```r
str(iris.hclust.fit)
```

### K-medoids

using; PAM algorithm (Partitioning Around Medoids, (Kaufman and Rousseeuw 1990))

Select a new point in each cluster that minimizes the sum of distances of all points in that cluster from itself.

### (T-SNE) t-distributed stochastic neighbor embedding

An ML approach for non-linear dimensionality reduction

It takes high dimensional data (many variables) and let's you visualize it in lower dimensions such as 2D

If you have groups that depend upon lots of variables and want to see which are closer together this method is useful

```r
#install.packages("Rtsne")
library('Rtsne')
```

```r
iris_unique = unique(iris)
iris_matrix = as.matrix(iris_unique[,1:4])
print( head( iris_matrix ) )
```

```r
tsne_out = Rtsne(iris_matrix,pca=FALSE,perplexity=10,theta=0.2)
names(tsne_out)
```

```r
#head( tsne_out$Y )
```

```r
plot(tsne_out$Y,col=iris_unique$Species,main='T-SNE of iris data')
```

```r
tsne_out = Rtsne(iris_unique,pca=FALSE, theta=0.0)

tsne_out = Rtsne(dist(normalize_input(iris_matrix)), theta=0.0)
plot(tsne_out$Y,col=iris_unique$Species, asp=1)
```

```r
tsne_out = Rtsne(iris_unique[,1:4],labels=as.factor(unique(iris[,5])))
print( head(tsne_out$Y) )

plot( (tsne_out$Y) ) 
```
